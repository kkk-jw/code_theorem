```python
import csv
f = open('C://Users//kjk53//OneDrive//바탕 화면//age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요:')
for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc("font", family = "Malgun Gothic")
plt.title(name + '지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
import csv
f = open('C://Users//kjk53//OneDrive//바탕 화면//age.csv')
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요: ')
for row in data:
    if name in row [0]:
        for i in row[3:103] :
            m.append(-int(i))
        for j in row[106:] :
            f.append(int(i))
            

import matplotlib.pyplot as plt
plt.figure(figsize = (10, 5), dpi=100)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역의 남녀 성별 인구 분포')
plt.barh(range(100), m, label = '남성')
plt.barh(range(100), f, label = '여성')
plt.legensd()
plt.show()
```
